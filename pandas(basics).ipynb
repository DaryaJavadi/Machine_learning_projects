```python
import pandas as pd
import numpy as np
```

```python
data = {
    "Song": ["Sunshine", "Moonlight", "Rainy Day", "Party All Night", "Love Story",
             "Hip Hop Beat", "Classical Calm", "Rock Anthem", "Pop Hit", "Summer Vibes"],
    "Artist": ["Alice", "Bob", "Charlie", "Alice", "Eve",
               "Frank", "Grace", "Henry", "Alice", "Bob"],
    "Genre": ["Pop", "Pop", "Jazz", "Pop", "Pop",
              "HipHop", "Classical", "Rock", "Pop", "Pop"],
    "StreamsMillions": [120, 250, np.nan, 180, 300,
                        90, 60, 200, np.nan, 150],
    "DurationSeconds": [210, np.nan, 300, 180, 220,
                        200, 360, 250, 230, np.nan],
    "ReleaseYear": [2021, 2020, 2019, 2021, 2018,
                    2020, 2017, 2019, 2021, 2022]
}
```

```python
# question 1:

df = pd.DataFrame(data)
df
```

```python
# question 2:

df.head()
```

```python
# question 3:

df.isna().sum()
```

```python
df['StreamsMillions'].fillna(0)
```

```python
df['DurationSeconds'].fillna(df['DurationSeconds'].mean())
```

```python
# question 4:

def categorize_streams(x):
    if x < 50:
        return "Low"
    elif x <= 200:
        return "Medium"
    else:
        return "High"

df['StreamCategory'] = df['StreamsMillions'].apply(categorize_streams)
df[['Song','StreamsMillions','StreamCategory']]
```

```python
# question 5:

df[(df['Genre']=='Pop') & (df['StreamCategory']=='High')]
```

```python
# question 6:

df.sort_values('StreamsMillions', ascending=False).head(3)
```

```python
# question 7:

artist_stats = df.groupby('Artist').agg({'StreamsMillions':'sum','DurationSeconds':'mean'})
artist_stats
```

```python
# question 8:

df.groupby('Genre')['StreamsMillions'].mean()
df.groupby('Genre').apply(lambda x: x.nlargest(2,'StreamsMillions'))
```

```python
# question 9:

df['HitSong'] = ((df['StreamsMillions'] > 150) & (df['DurationSeconds'] < 240))
df[['Song','StreamsMillions','DurationSeconds','HitSong']]
```
