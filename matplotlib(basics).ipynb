```python
import kagglehub

path = kagglehub.dataset_download("heesoo37/120-years-of-olympic-history-athletes-and-results")

print("Path to dataset files:", path)
```

```python
import pandas as pd

df = pd.read_csv(path + "/athlete_events.csv")
```

```python
# step 1:

events_per_year = df.groupby("Year")["Event"].nunique()

import matplotlib.pyplot as plt
plt.figure()
plt.plot(events_per_year.index, events_per_year.values)
plt.xlabel("Year")
plt.ylabel("Number of Events")
plt.title("Number of Olympic Events Over Time")
plt.show()
```

```python
# step 2:

athletes_per_year = df.groupby("Year")["ID"].nunique()

plt.figure()
plt.plot(athletes_per_year.index, athletes_per_year.values, color = "red")
plt.xlabel("Year")
plt.ylabel("Number of Athletes")
plt.title("Athlete Participation Over Time")
plt.grid(True)
plt.show()
```

```python
# step 3:

gender_counts = df["Sex"].value_counts()

plt.figure()
plt.bar(gender_counts.index, gender_counts.values)
plt.xlabel("Gender")
plt.ylabel("Number of Athletes")
plt.title("Gender Participation")
plt.show()
```

```python
# step 4:

medals = df.dropna(subset=["Medal"])
country_medals = medals["NOC"].value_counts().head(10)

plt.figure()
plt.barh(country_medals.index, country_medals.values)
plt.xlabel("Medal Count")
plt.title("Top 10 Countries by Medals")
plt.gca().invert_yaxis()
plt.show()
```

```python
# step 5:

medal_counts = medals["Medal"].value_counts()

plt.figure()
plt.pie(medal_counts.values, labels=medal_counts.index, autopct="%1.1f%%")
plt.title("Medal Distribution")
plt.show()
```

```python
# step 6:

ages = df["Age"].dropna()

plt.figure()
plt.hist(ages, bins=3, color = 'skyblue', edgecolor = 'black')
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Athlete Age Distribution")
plt.show()
```

```python
# step 7:

medal_map = {"Bronze": 1, "Silver": 2, "Gold": 3}
medals_age = medals.dropna(subset=["Age", "Medal"])
y = medals_age["Medal"].map(medal_map)

plt.figure()
plt.scatter(medals_age["Age"], y)
plt.xlabel("Age")
plt.ylabel("Medal Type")
plt.title("Age vs Medal")
plt.show()
```

```python
# step 8:

hw = df.dropna(subset=["Height", "Weight"])

plt.figure()
plt.scatter(hw[hw["Sex"]=="M"]["Height"], hw[hw["Sex"]=="M"]["Weight"], marker="o", label="Male")
plt.scatter(hw[hw["Sex"]=="F"]["Height"], hw[hw["Sex"]=="F"]["Weight"], marker="^", label="Female")
plt.xlabel("Height")
plt.ylabel("Weight")
plt.title("Height vs Weight by Gender")
plt.legend()
plt.show()
```

```python
# step 9:

top_sports = df["Sport"].value_counts().head(15)

plt.figure()
plt.bar(top_sports.index, top_sports.values)
plt.xticks(rotation=90)
plt.ylabel("Number of Athletes")
plt.title("Top 15 Sports")
plt.show()
```

```python
# step 10:

season_counts = df["Season"].value_counts()

plt.figure()
plt.bar(season_counts.index, season_counts.values)
plt.xlabel("Season")
plt.ylabel("Number of Athletes")
plt.title("Summer vs Winter Olympics")
plt.show()
```

```python
# step 11:

usa = medals[medals["NOC"] == "USA"]
usa_medals_year = usa.groupby("Year")["Medal"].count()

plt.figure()
plt.plot(usa_medals_year.index, usa_medals_year.values)
plt.xlabel("Year")
plt.ylabel("Medals")
plt.title("USA Medals Over Time")
plt.show()
```

```python
# step 12:

plt.figure(figsize=(10,8))

plt.subplot(2,2,1)
plt.plot(athletes_per_year.index, athletes_per_year.values)
plt.title("Athletes per Year")

plt.subplot(2,2,2)
plt.pie(medal_counts.values, labels=medal_counts.index)

plt.subplot(2,2,3)
plt.bar(gender_counts.index, gender_counts.values)
plt.title("Gender Count")

plt.subplot(2,2,4)
plt.bar(top_sports.index, top_sports.values)
plt.xticks(rotation=90)
plt.title("Top Sports")

plt.tight_layout()
plt.show()
```
