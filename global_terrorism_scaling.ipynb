```python
df = pd.read_csv('/content/globalterrorismdb_0718dist.csv', encoding="latin1")
df
```

```python
# 1.
print("Rows:", df.shape[0])
print("Columns:", df.shape[1])
```

```python
# 2.
missing = df.isnull().sum().sort_values(ascending=False)
print(missing.head(20))
```

```python
# 3.
numeric_cols = df.select_dtypes(include=["int64","float64"]).columns
categorical_cols = df.select_dtypes(include=["object"]).columns

print("Numeric:", len(numeric_cols))
print("Categorical:", len(categorical_cols))
```

```python
# 4.
top_countries = df["country_txt"].value_counts().head(10)
print(top_countries)
```

```python
# 5.
top_attack = df["attacktype1_txt"].value_counts().head(1)
print(top_attack)
```

```python
# 6.
def outlier_count(col):
    Q1 = col.quantile(0.25)
    Q3 = col.quantile(0.75)
    IQR = Q3 - Q1
    return ((col < Q1 - 1.5*IQR) | (col > Q3 + 1.5*IQR)).sum()

numeric_cols = df.select_dtypes(include=["int64", "float64"])
outliers = numeric_cols.apply(outlier_count)

top5 = outliers.sort_values(ascending=False).head(5)
print(top5)
```

```python
# 7.
nkill_outliers = outlier_count(df["nkill"].dropna())
nwound_outliers = outlier_count(df["nwound"].dropna())

print("nkill outliers:", nkill_outliers)
print("nwound outliers:", nwound_outliers)
```

```python
# 8.
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_std = scaler.fit_transform(df[["nkill","nwound"]].fillna(0))

df_std = pd.DataFrame(scaled_std, columns=["nkill_std", "nwound_std"])
print(df_std.head())
```

```python
# 9.
print("Mean:", df_std.mean())
print("Std:", df_std.std())
```

```python
# 10.
from sklearn.preprocessing import MinMaxScaler

minmax = MinMaxScaler()
scaled_mm = minmax.fit_transform(df[["nkill","nwound"]].fillna(0))

df_mm = pd.DataFrame(scaled_mm, columns=["nkill_mm", "nwound_mm"])
print(df_mm.head())
```

```python
# 11.
print("Min:", df_mm.min())
print("Max:", df_mm.max())
```

```python
# 12.
from sklearn.preprocessing import RobustScaler

robust = RobustScaler()
scaled_rb = robust.fit_transform(df[["nkill","nwound"]].fillna(0))

df_rb = pd.DataFrame(scaled_rb, columns=["nkill_rb", "nwound_rb"])
print(df_rb.head())
```

```python
# 13.
print("StandardScaler mean:", df_std.mean().values)
print("StandardScaler std:", df_std.std().values)

print("MinMaxScaler min:", df_mm.min().values)
print("MinMaxScaler max:", df_mm.max().values)

print("RobustScaler median:", df_rb.median().values)
print("RobustScaler IQR:", (df_rb.quantile(0.75) - df_rb.quantile(0.25)).values)
```
